```python
#encoding
#GET_DUMMIES

df2=pd.get_dummies(df1[['Location','Fuel_Type','Transmission','Owner_Type']],drop_first=True)
df2
```

```python
#concat
dfe=pd.concat([df1,df2],axis=1)
dfe
```

```python
dfe.dtypes
```

```python
#drop already encoded columns , name , index & new price column

dfe.drop(['Unnamed: 0','Name','Location','Fuel_Type','Transmission','Owner_Type','New_Price'],axis=1,inplace=True)
dfe
```

```python
dfe.dtypes
```

```python
#STRING CONVERSION
#removing the units in 'mileage','engine' and 'power' to convert from object dtype to int
#for eg: replacing km/kg to ''(non-space)
#replace 'null' to '0'

dfe['Mileage']=dfe['Mileage'].str.replace('km/kg','')
dfe['Mileage']=dfe['Mileage'].str.replace('kmpl','')
dfe['Engine']=dfe['Engine'].str.replace('CC','')
dfe['Power']=dfe['Power'].str.replace('bhp','')

dfe['Mileage']=dfe['Mileage'].str.replace('null','0')
dfe['Engine']=dfe['Engine'].str.replace('null','0')
dfe['Power']=dfe['Power'].str.replace('null','0')

dfe
```

```python
dfe.dtypes
```

```python
#DTYPE CONVERSION

dfe['Engine']=dfe['Engine'].astype(float)
dfe['Mileage']=dfe['Mileage'].astype(float)
dfe['Power']=dfe['Power'].astype(float)
```

```python
dfe.dtypes
```

```python
#convert '0's(replacements of null) to null value                    NaN=not a number

dfe.loc[dfe.Engine==0,'Engine']=np.NaN
dfe.loc[dfe.Mileage==0,'Mileage']=np.NaN
dfe.loc[dfe.Power==0,'Power']=np.NaN
```

```python
dfe.isna().sum()
```

```python
#fill missing values

dfe['Mileage']=dfe['Mileage'].fillna(dfe['Mileage'].mean())
dfe['Engine']=dfe['Engine'].fillna(dfe['Engine'].mean())
dfe['Power']=dfe['Power'].fillna(dfe['Power'].mean())

dfe['Seats']=dfe['Seats'].fillna(dfe['Seats'].mode()[0])

dfe.isna().sum()
```

```python
#drop fuel electric column bcz it is not present in testing data

dfe.drop(['Fuel_Type_Electric'],axis=1,inplace=True)
```

```python
#separation of x & y

x=dfe.drop(['Price'],axis=1)
x
```
